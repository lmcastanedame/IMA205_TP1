```python
def pcaLecture(X):
    ''' 
    Inputs: 
            X: is a [Nxd] matrix. Every row is an observation and every
              column consists of features.
    Outputs:
            Y: is a [Nxd] matrix representing the scores, namely the 
            coordinates of X onto the new basis given by the eigenvactors U
            of the covariance matrix of X. Columns are the principal components.
               
            U: columns are Eigenvectors (sorted from the greatest to the lowest eigenvalue)
    
            D: Eigenvalues (sorted from the greatest to the lowest eigenvalue)
               
            var_explained: percentage of the original variability explained
            by each principal component.
    '''
    
    N = X.shape[0]

    # centering
    Xc = X - np.mean(X, axis=0)

    # computation of eigenvectors and eigenvalues using SVD
    D2, Uh = LA.svd(Xc, full_matrices=False)[1:]
    U = Uh.T

    # computation of the scores
    Y = np.dot(Xc,U)

    # computation of the eigenvalues
    D = D2**2/(N-1) 
    tot = np.sum(D)

    # computation of explained variance
    var_explained = D/tot
    return Y,U,D,var_explained
```

- Kernel PCA is an extension of Principal Component Analysis (PCA) that uses techniques known as the "kernel trick" to enable PCA to be applied to data that is not linearly separable. This is particularly useful when the relationship between variables in the data is nonlinear. The kernel trick involves mapping the original data into a higher-dimensional space where the nonlinear relationship can be represented as linear, allowing standard PCA to be applied.

    1. **Nonlinear Dimensionality Reduction**: Kernel PCA can capture complex, nonlinear relationships within the data, whereas standard PCA is limited to linear relationships.

    2. **Flexibility**: Different kernel functions (e.g., polynomial, radial basis function, sigmoid) can be used in Kernel PCA to best capture the structure of the data. This flexibility allows Kernel PCA to be tailored to the specific nonlinearities in the dataset.

    3. **Improved Feature Separation**: By mapping data to a higher-dimensional space, Kernel PCA can make it easier to separate data points that are not linearly separable in the original feature space, which can be beneficial for classification and clustering tasks.

```python
def KpcaGaussianLecture(X,gamma):
    '''  
    Inputs: 
            X: is a [Nxd] matrix. Every row is an observation and every
            column is a feature.
 
    Outputs:
            Y: is a [Nxd] matrix representing the scores, namely the 
            coordinates of \phi(X) onto the new basis given by the eigenvactors 
            of the covariance matrix of \phi(X). Columns are the principal components.

            An: columns are Eigenvectors normalised (sorted from the greatest
            to the lowest eigenvalue)
     
            D: Eigenvalues (sorted from the greatest to the lowest eigenvalue)

            var_explained: percentage of the original variability explained
            by each principal component.

    '''

    # kernel matrix using a rbf kernel 
    def kernel_matrix(X,gamma):
        N = X.shape[0]
        InnerX = np.dot(X,X.T)
        temp1 = np.sum(X**2,axis=1).reshape((N,1))
        temp2 = np.sum(X**2,axis=1).reshape((1,N))
        Norm1 = np.repeat(temp1,N,axis=1)
        Norm2 = np.repeat(temp2,N,axis=0) 
        Norm = Norm1+Norm2-2*InnerX;
        Norm[Norm<1e-10] = 0;    
        K = np.exp(-gamma*Norm)
        return K
       
    N = X.shape[0]

    K = kernel_matrix(X,gamma) # kernel matrix
    
    # center kernel matrix
    oneN = np.ones((N,N))/N;
    # center kernel matrix
    Kc = K - np.dot(oneN,K) - np.dot(K,oneN) + np.dot(np.dot(oneN,K),oneN)

    # eigenvalue analysis
    D,A = LA.eigh(Kc)     
    idx = D.argsort()[::-1]  # reverse order to make 'descend' 
    D = np.real(D[idx])
    D[D<0] = 1e-18 # make negative eigenvalues positive (and almost 0)
    A = np.real(A[:,idx])

    # variance explained
    tot = np.sum(D)

    # computation of explained variance
    var_explained = D/tot

    # Normalisation eigenvectors
    # Norm of every eigenvector is 1, we want it to be 1/sqrt(N*eig)
    
    An = np.copy(A)
    for i in range(N):      
        An[:,i] = np.dot(A[:,i],(1/np.sqrt((N-1)*D[i])) )

    # computation of the scores    
    Y = np.dot(Kc,An)
    
    return Y,An,D,var_explained
```

Independent Component Analysis (ICA) is a computational method for separating a multivariate signal into additive subcomponents that are maximally independent. It is commonly used in scenarios where the goal is to identify underlying factors or sources from mixed signals. Unlike PCA, which focuses on maximizing variance and finding orthogonal components, ICA aims to find components that are statistically independent from each other.

ICA can work better than PCA in the following scenarios:

1. **When the Data Contains Mixed Signals**: ICA excels in situations where the observed data are mixtures of unknown latent variables (or sources), and the goal is to separate these mixtures into their constituent sources.

2. **When Independence is More Important Than Variance**: In cases where signals or features are independent but not necessarily orthogonal or uncorrelated, ICA can separate these sources more effectively than PCA.

- **Scenario 1**: ICA appears to have separated the two classes along the first independent component reasonably well, suggesting that the underlying sources of the signals (features) are independent. If these sources correspond to our class separation, then ICA would be working effectively here.

- **Scenario 2**: the first independent component shows overlapping scores for Class 0 and Class 1, it suggests that this component has not captured features that are independent in a way that also corresponds to the class distinction. ICA would not be deemed successful in this scenario for the purpose of class separation.

- **Scenario 3**: The classes are distributed in a way that suggests there might be underlying independent processes or factors at play. If the ICA components correspond to these factors, ICA works better than PCA by isolating these independent sources even if they don't correspond to maximum variance.

- **Scenario 4**: This scenario seems to depict a nonlinear relationship between classes, likely unsolvable by linear methods like PCA. If the circular components correspond to independent sources of variance in the data, ICA might be able to separate them effectively, which PCA could not.

In general, ICA will outperform PCA when the assumption of statistical independence is more appropriate for the data at hand than the assumption of uncorrelated components. It's important to note that ICA does not provide components in order of importance (like PCA does with variance), and all components are considered equally important. The success of ICA is also highly dependent on the assumption that the sources are statistically independent and that the mixing system is linear and instantaneous.

```python
def FastICALecture(X,N_Iter=3000,tol=1e-5,plot_evolution=0):
    '''
    Inputs: 
                   X: is a [d x N] matrix. Every column is an observation 
                   and every row is a feature.       
    
                   (Optional) N_Iter: maximum number of iterations
    
                   (Optional) delta: convergence criteria threshold
    
                   (Optional) plot_evolution: plot evolution of error
                   
     Outputs:      
                   S: [d x N] matrix.  Each column is an independent component 
                   of the centred and whitened input data X              
                   
                   W: [d x d] matrix. It is the demixing matrix. S = W*Xcw 
    
    '''
    
    # We choose G =-exp(-1/2t^2)
    # First derivative of G
    def g(t):
        res = t*np.exp(-(1/2)*(t**2))
        return res
    
    # Second derivative of G  
    def gp(t):
        res = (1 - t**2) * np.exp(-(1/2) * (t**2))
        return res
    
    # Size of X
    d,N = X.shape 
        
    # Compute sample mean
    mu = np.mean(X,axis=1)
    
    # Center data
    Xc = X - mu[:, np.newaxis]

    # Compute covariance matrix
    C = np.cov(Xc, rowvar=True)
    
    # Whiten data
    Xcw = np.dot(np.linalg.inv(np.linalg.cholesky(C)).T, Xc)
    
    # check if are whitened
    if np.sum(np.eye(d) - np.abs(np.cov(Xcw)))>1e-10:
        raise NameError('Your whitening transformation does not work...')
    
    # Initialize W
    W = ortho_group.rvs(d) # random orthogonal matrix 
    
    # delta evolution
    k = 0
    delta = np.inf
    evolutionDelta=[]
    
    while delta > tol and k < N_Iter:
    
        k = k + 1
        W_old = np.copy(W)
        
        # Update rule for W
        Wp = np.dot(g(np.dot(W, Xcw)), Xcw.T) - np.diag(np.dot(gp(np.dot(W, Xcw)), np.ones(N))) @ W
        W = np.linalg.qr(Wp)[0]  # Enforce orthogonality using QR decomposition
        # W = LA.inv( np.sqrt( Wp @ Wp.T ) ) @ Wp # Enforce orthogonality using normalization

        if np.sum(np.eye(d) - np.abs(np.dot(W,W.T)))>1e-10:
            raise NameError('W should be an orthogonal matrix. Check the computations')
                 
        delta = 1 - np.min( np.abs( np.diag( np.dot(W.T,W_old) ) ) )
        evolutionDelta.append(delta)
        
        if k == 1 or k % 100 == 0:
            print('Iteration ICA number ', k, ' out of ', N_Iter , ', delta = ', delta)
     
        
    if k == N_Iter:
        print('Maximum number of iterations reached ! delta = ', delta)
    else:
        print('Convergence achieved ( delta = ', delta, ') in ', k, ' iterations')


    # Independent components
    S = np.dot(W, Xcw)
            
    if plot_evolution == 1:
        plt.figure(figsize=(6, 6))
        plt.plot(range(k),evolutionDelta,'bx--', linewidth=4, markersize=12)  
        plt.title('Evolution of error - ICA')
        plt.show()
       
    return S, W, Xcw
```

**Theoretical Questions:** 

1. In your opinion, why do we need to center the data before computing a PCA ? If you want, you can use the previous toy examples to answer this question.

Centering the data before performing Principal Component Analysis (PCA) is a critical step for ensuring the algorithm works correctly and yields meaningful results. 

PCA aims to identify the directions (principal components) along which the variance in the data is maximized. By centering the data (subtracting the mean of each variable), the PCA model finds components that account for variance relative to the mean of the data. If the data were not centered, the first principal component might instead reflect the mean of the data rather than variations around the mean, which is usually more interesting.

PCA often involves the eigen decomposition of the covariance matrix of the data or the singular value decomposition (SVD) of the data matrix itself. The covariance between two variables is influenced by their respective means. Centering ensures that the covariance matrix reflects covariances in fluctuations around a central point (the mean), which is crucial for finding directions of maximal variance.

From a geometric perspective, centering the data moves the origin of the coordinate system to the mean of the data. This makes the principal components represent directions of variance from the central point (the mean of the data). Without centering, these directions could be skewed by the location of the data relative to the origin of the coordinate system, leading to misleading interpretations.

2. Let $X$ be the original data, a matrix $\left[ N,d \right]$, and $Y$ the scores of a PCA keeping all eigenvectors, which means that $Y$ is also a matrix $\left[ N,d \right]$. Are $X$ and $Y$ equal ? If not, why ? What would you use (generally speaking) in a machine learning problem ? Why ?

When performing Principal Component Analysis (PCA) the resulting scores matrix $Y$ (of size $[N, d]$) is not equal to the original data matrix $X$, despite having the same dimensions. PCA transforms the original data into a new coordinate system where the axes (principal components) are orthogonal and represent the directions of maximum variance in the data. The scores matrix $Y$ represents the coordinates of the original data in this new space. Even if all eigenvectors are kept, the basis of the space has changed; $X$ is represented in the original feature space, while $Y$ is represented in the principal component space.

The process of PCA can be thought of as rotating the dataset around the origin of the feature space to align with the principal components. This rotation changes the representation of the data points, although the relative distances and relationships between them are preserved in terms of variance and covariance.

In many machine learning problems, especially those involving high-dimensional data, PCA is used for feature reduction to improve model performance, reduce overfitting, and decrease computational costs. By selecting a subset of components that capture most of the variance, we can work with a lower-dimensional representation ($Y$) that retains the most informative aspects of the original data ($X$).

If the primary goal is dimensionality reduction or capturing latent structures in the data, we would use the scores ($Y$). This is particularly useful in unsupervised learning, exploratory data analysis, or preprocessing steps before supervised learning.

If the model benefits from the original features and there's no need to reduce dimensionality, we might prefer to use the original data ($X$).

3. Let $x_p$ and $x_q$ be two row-vectors representing two images, $U$ an *orthogonal* matrix whose columns are the eigenvectors of $X$ and $y_p=x_pU$, $y_q=x_qU$, check that $x_px_q^T=y_py_q^T$. This shows that $Y=XU$ is a linear transformation that preserves inner products.

To show that $x_p x_q^T = y_p y_q^T$ given $y_p = x_p U$ and $y_q = x_q U$, where $U$ is an orthogonal matrix whose columns are the eigenvectors of $X$, we need to expand the expression and use properties of orthogonal matrices. An orthogonal matrix $U$ has the property that $U^T U = U U^T = I$, where $I$ is the identity matrix, and this implies that $U^{-1} = U^T$.

Given:
- $y_p = x_p U$
- $y_q = x_q U$

We want to show:
- $x_p x_q^T = y_p y_q^T$

### Proof:

Starting from the right-hand side:

$$y_p y_q^T = (x_p U) (x_q U)^T$$

Since $(AB)^T = B^T A^T$, we apply this property:

$$= (x_p U) (U^T x_q^T)$$

Because matrix multiplication is associative, we can rearrange the parentheses:

$$= x_p (U U^T) x_q^T$$

Given $U$ is orthogonal, $U U^T = I$ where $I$ is the identity matrix, so:

$$= x_p I x_q^T$$

The identity matrix does not change the vector it multiplies, so:

$$= x_p x_q^T$$

This shows that $x_p x_q^T = y_p y_q^T$, which means that the transformation $Y = XU$ preserves the inner products between the vectors in $X$. This property is fundamental in many applications, including dimensionality reduction and data analysis, because preserving inner products means preserving angles and lengths between vectors, thus maintaining the geometric and statistical relationships between data points in the transformed space.

4. Let $C$ be the covariance matrix of $X$ and $C=UDU^T$ its eigen decomposition. Show that the covariance matrix of $Y=XU$ is $D$.

The covariance matrix $C$ is defined as $C = \frac{1}{n-1} (X - \bar{X})^T (X - \bar{X})$, where $\bar{X}$ is the mean of $X$, and $n$ is the number of observations. Assuming $X$ has already been mean-centered:

Given:
- $Y = XU$
- $C = UDU^T$ is the eigen decomposition of $C$, where $C$ is the covariance matrix of $X$, $U$ is the matrix of eigenvectors, and $D$ is the diagonal matrix of eigenvalues.

### Proof:

1. **Covariance Matrix of $Y$**: The covariance matrix of $Y$ is given by $C_Y = \frac{1}{n-1}Y^TY$.

2. **Substitute $Y = XU$**: 
   $$C_Y = \frac{1}{n-1}(XU)^T(XU)$$

3. **Apply Matrix Transpose Properties**: 
   $$C_Y = \frac{1}{n-1}U^TX^TXU$$

4. **Substitute $C$ for $\frac{1}{n-1}X^TX$**:
   Since $C = \frac{1}{n-1}X^TX$, we can replace $\frac{1}{n-1}X^TX$ with $C$:
   $$C_Y = U^TCU$$

5. **Substitute $C = UDU^T$**:
   $$C_Y = U^T(UDU^T)U$$

6. **Apply Matrix Multiplication**:
   Since matrix multiplication is associative, $U^T(UDU^T)U = (U^TU)D(U^TU)$.

7. **Orthogonal Eigenvectors**:
   Given $U$ is orthogonal, $U^TU = I$, where $I$ is the identity matrix:
   $$C_Y = IDI = D$$

The covariance matrix of $Y = XU$ is $D$, the diagonal matrix of eigenvalues from the eigen decomposition of the covariance matrix $C$ of $X$. This result shows that when we transform the original data $X$ using its eigenvectors $U$, the covariance matrix of the transformed data $Y$ directly corresponds to the diagonal matrix of eigenvalues $D$, which indicates that the transformed features in $Y$ are uncorrelated and their variances are the eigenvalues in $D$.

3. Why the basis vectors $\{  \mathbf{\alpha}_i \}$ are not plotted as in PCA ?

- The basis vectors (or principal components) in Kernel PCA (KPCA) are fundamentally different from those in linear PCA, which makes their direct visualization and interpretation more challenging. This difference stems from the nature of the feature space transformation and the computation of these basis vectors. Here are the key reasons:

    1. **High-dimensional (or Infinite) Feature Space**: KPCA implicitly maps the input data into a higher-dimensional feature space (possibly infinite-dimensional with certain kernels like the RBF kernel) through the use of kernel functions. The basis vectors $\{\mathbf{\alpha}_i\}$ in KPCA lie in this high-dimensional feature space, not in the original input space. In contrast, PCA operates directly in the original feature space, making its principal components (which are linear combinations of the original features) directly interpretable and plottable in the context of the original data.

    2. **Implicit Feature Space**: The mapping to the feature space in KPCA is implicit, meaning we never compute the coordinates of the data points in the feature space directly. Instead, the kernel function computes the inner products between the images of the data points in the feature space. Because we never explicitly know the coordinates in the feature space, we cannot directly visualize the basis vectors $\{\mathbf{\alpha}_i\}$ as we do with PCA, where the components are explicit linear combinations of the original features.

    3. **Relational Nature of Basis Vectors**: In KPCA, the basis vectors $\{\mathbf{\alpha}_i\}$ represent directions in the feature space that maximize variance based on the kernel-induced inner product. However, these directions do not have a straightforward interpretation in the original input space. They can be thought of more as relations between data points as captured by the kernel function, rather than explicit directions or axes along which the data varies.

    4. **Visualization through Backprojection**: While direct visualization of $\{\mathbf{\alpha}_i\}$ in KPCA is not feasible, some approaches attempt to visualize the effects of KPCA transformations. One method involves backprojecting the reduced-dimensionality data points to the original space (if an inverse transform is possible or can be approximated). This backprojection can offer some intuition about the data's structure in the reduced space, but it still doesn't visualize the basis vectors directly.

### Performance

- **PCA/KPCA**: Both PCA and KPCA can significantly improve the performance of machine learning models by reducing dimensionality, thereby mitigating the curse of dimensionality. By focusing on the components that account for the most variance, they can enhance the signal-to-noise ratio in the data. KPCA, with its ability to capture nonlinear relationships, can further uncover structure in the data that linear methods like PCA might miss.
- **Original Pixel Intensities**: Using raw pixel intensities might lead to poor performance for high-dimensional data due to the curse of dimensionality. Models trained on raw pixels are prone to overfitting and might require significantly more data to achieve comparable performance.

### Computational Time

- **PCA**: PCA is computationally efficient, especially for linear dimensionality reduction. It's well-suited for large datasets and can be easily scaled.
- **KPCA**: KPCA is more computationally intensive than PCA due to the need to compute the kernel matrix and potentially deal with a high-dimensional feature space. The computational complexity can become a bottleneck for very large datasets.
- **Original Pixel Intensities**: Using original pixel intensities eliminates the preprocessing step, thus saving time initially. However, the subsequent training time for machine learning models might increase significantly, as these models have to deal with a larger number of features.

### Number of Features

- **PCA/KPCA**: Both methods reduce the number of features needed to represent the data effectively, which can simplify models and reduce the risk of overfitting. This reduction also makes models faster to train and can improve their generalization capabilities.
- **Original Pixel Intensities**: Working with original pixel intensities means dealing with all available features, which can be prohibitively large for image data. This can increase the computational burden and memory requirements for training models.

### Interpretability of the Results

- **PCA**: PCA provides components that are linear combinations of the original features, which can sometimes be interpreted (e.g., in terms of which original features contribute most to each principal component). However, this interpretability is often challenging in the context of images, where components do not necessarily correspond to intuitive visual patterns.
- **KPCA**: KPCA's nonlinear components make it even more challenging to interpret the results in terms of the original features. The transformation into a high-dimensional space, while powerful for capturing complex patterns, abstracts away from the original input space.
- **Original Pixel Intensities**: Using raw pixel values retains the most direct interpretation of the data, as each feature corresponds to a specific pixel's intensity. This can be advantageous for tasks where the spatial arrangement and exact values of pixels are critical for interpretation.

```python
# New funtion Kpca_poly_lecture, with kernerl k(x, y) = < x, y, >^d
def KpcaPolyLecture(X,d):
    '''  
    Inputs: 
            X: is a [Nxd] matrix. Every row is an observation and every
            column is a feature.
 
            d: degree of the polynomial kernel
    
    Outputs:
            Y: is a [Nxd] matrix representing the scores, namely the 
            coordinates of \phi(X) onto the new basis given by the eigenvactors 
            of the covariance matrix of \phi(X). Columns are the principal components.

            An: columns are Eigenvectors normalised (sorted from the greatest
            to the lowest eigenvalue)
     
            D: Eigenvalues (sorted from the greatest to the lowest eigenvalue)

            var_explained: percentage of the original variability explained
            by each principal component.

    '''

    # kernel matrix using a polynomial kernel 
    def kernel_matrix(X,d):
        N = X.shape[0]
        K = np.dot(X,X.T)
        K = (1 + K)**d
        return K
       
    N = X.shape[0]

    K = kernel_matrix(X,d) # kernel matrix
    
    # center kernel matrix
    oneN = np.ones((N,N))/N;
    # center kernel matrix
    Kc = K - np.dot(oneN,K) - np.dot(K,oneN) + np.dot(np.dot(oneN,K),oneN)

    # eigenvalue analysis
    D,A = LA.eigh(Kc)     
    idx = D.argsort()[::-1]  # reverse order to make 'descend' 
    D = np.real(D[idx])
    D[D<0] = 1e-18 # make negative eigenvalues positive (and almost 0)
    A = np.real(A[:,idx])

    # variance explained
    tot = np.sum(D)

    # computation of explained variance
    var_explained = D/tot

    # Normalisation eigenvectors
    # Norm of every eigenvector is 1, we want it to be 1/sqrt(N*eig)
    
    An = np.copy(A)
    for i in range(N):      
        An[:,i] = np.dot(A[:,i],(1/np.sqrt((N-1)*D[i])) )

    # computation of the scores    
    Y = np.dot(Kc,An)
    
    return Y,An,D,var_explained
```

# II - 3) ICA (optional part)

In the next section you will evaluate ICA. Every image $x_i$ can be seen as a linear combination of basis images. ICA can be used in two different ways for face recognition. We can look for a set of statistically independent basis images $s_j$ (first architecture) or for a set of statistically independent coefficients $a_{j}$ (second architecture).

In the first architecture, we compute $X'=A'S'$, where every row of $X'$ is an image and the columns are pixels. Images are considered as random variables and we look for a set of statistically independent basis images contained in the rows of $S'$.

In the second architecture, we transpose the previous setting computing $X''=A''S''$, where every column of $X''$ is an image and rows are pixels. In this case, we consider the pixels as random variables and we look for a set of statistically independent coefficients contained in the rows of $S$ and a set of basis images in the columns of $A$.

Instead than using the original training data $X$ as input matrix, we are going to use the eigenvectors (first architecture) or the scores (second architecture) computed with PCA, namely $Y=XL$ (same notation as in the slides of the lecture). In this way, we reduce the computational time since the number of eigenvectors that account for 99\% of the variance of the training images (columns of $L$) is definitely lower than the number of pixels (columns of $X$). If you want, you can of course use the original data but it will take much more time to converge.

For the first architecture we will use $L^T$ as input matrix. In fact, we can notice that the PCA approximation of the matrix $X_{train}$, containing an image in every row, can be written as $\tilde{X} = YL^T$. If we use $L^T$ as input in the ICA algorithm we obtain $L^T=AS$, thus it follows that $\tilde{X}=YW^TS$ (since $A=W^{-1}=W^T$). The basis images are contained in the rows of $S$ and the coefficients used for evaluating the performance are instead contained in the rows of $Y_{train}W^T$ for the training set and in $Y_{test}W^T$ for the test set.

For the second architecture, we will instead use $Y^T$ as input matrix thus obtaining $Y^T=AS$. Remember that in the second architecture we want to apply the ICA algorithm to the transpose of $X_{train}$. We can notice that, given the PCA transformation $Y=XL$, one can write $X \approx YL^T$ which entails $X^T \approx LY^T=LAS=LW^TS$. The columns of $LW^T$ contain the basis images whereas the columns of $S$ contain the statistically independent coefficients used to test the performance of the algorithm. The coefficients for the test set are in the columns of $S_{test}=W_{train}Y_{test}^T$.
