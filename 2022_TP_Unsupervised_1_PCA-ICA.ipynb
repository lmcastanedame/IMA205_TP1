```python
def pcaLecture(X):
    ''' 
    Inputs: 
            X: is a [Nxd] matrix. Every row is an observation and every
              column consists of features.
    Outputs:
            Y: is a [Nxd] matrix representing the scores, namely the 
            coordinates of X onto the new basis given by the eigenvactors U
            of the covariance matrix of X. Columns are the principal components.
               
            U: columns are Eigenvectors (sorted from the greatest to the lowest eigenvalue)
    
            D: Eigenvalues (sorted from the greatest to the lowest eigenvalue)
               
            var_explained: percentage of the original variability explained
            by each principal component.
    '''
    
    N = X.shape[0]

    # centering
    Xc = X - np.mean(X, axis=0)

    # computation of eigenvectors and eigenvalues using SVD
    D2, Uh = LA.svd(Xc, full_matrices=False)[1:]
    U = Uh.T

    # computation of the scores
    Y = np.dot(Xc,U)

    # computation of the eigenvalues
    D = D2**2/(N-1) 
    tot = np.sum(D)

    # computation of explained variance
    var_explained = D/tot
    return Y,U,D,var_explained
```

1. What is the interest of kernel PCA over the basic PCA?

- Kernel PCA is an extension of Principal Component Analysis (PCA) that uses techniques known as the "kernel trick" to enable PCA to be applied to data that is not linearly separable. This is particularly useful when the relationship between variables in the data is nonlinear. The kernel trick involves mapping the original data into a higher-dimensional space where the nonlinear relationship can be represented as linear, allowing standard PCA to be applied.

    1. **Nonlinear Dimensionality Reduction**: Kernel PCA can capture complex, nonlinear relationships within the data, whereas standard PCA is limited to linear relationships.

    2. **Flexibility**: Different kernel functions (e.g., polynomial, radial basis function, sigmoid) can be used in Kernel PCA to best capture the structure of the data. This flexibility allows Kernel PCA to be tailored to the specific nonlinearities in the dataset.

    3. **Improved Feature Separation**: By mapping data to a higher-dimensional space, Kernel PCA can make it easier to separate data points that are not linearly separable in the original feature space, which can be beneficial for classification and clustering tasks.

```python
def KpcaGaussianLecture(X,gamma):
    '''  
    Inputs: 
            X: is a [Nxd] matrix. Every row is an observation and every
            column is a feature.
 
    Outputs:
            Y: is a [Nxd] matrix representing the scores, namely the 
            coordinates of \phi(X) onto the new basis given by the eigenvactors 
            of the covariance matrix of \phi(X). Columns are the principal components.

            An: columns are Eigenvectors normalised (sorted from the greatest
            to the lowest eigenvalue)
     
            D: Eigenvalues (sorted from the greatest to the lowest eigenvalue)

            var_explained: percentage of the original variability explained
            by each principal component.

    '''

    # kernel matrix using a rbf kernel 
    def kernel_matrix(X,gamma):
        N = X.shape[0]
        InnerX = np.dot(X,X.T)
        temp1 = np.sum(X**2,axis=1).reshape((N,1))
        temp2 = np.sum(X**2,axis=1).reshape((1,N))
        Norm1 = np.repeat(temp1,N,axis=1)
        Norm2 = np.repeat(temp2,N,axis=0) 
        Norm = Norm1+Norm2-2*InnerX;
        Norm[Norm<1e-10] = 0;    
        K = np.exp(-gamma*Norm)
        return K
       
    N = X.shape[0]

    K = kernel_matrix(X,gamma) # kernel matrix
    
    # center kernel matrix
    oneN = np.ones((N,N))/N;
    # center kernel matrix
    Kc = K - np.dot(oneN,K) - np.dot(K,oneN) + np.dot(np.dot(oneN,K),oneN)

    # eigenvalue analysis
    D,A = LA.eigh(Kc)     
    idx = D.argsort()[::-1]  # reverse order to make 'descend' 
    D = np.real(D[idx])
    D[D<0] = 1e-18 # make negative eigenvalues positive (and almost 0)
    A = np.real(A[:,idx])

    # variance explained
    tot = np.sum(D)

    # computation of explained variance
    var_explained = D/tot

    # Normalisation eigenvectors
    # Norm of every eigenvector is 1, we want it to be 1/sqrt(N*eig)
    
    An = np.copy(A)
    for i in range(N):      
        An[:,i] = np.dot(A[:,i],(1/np.sqrt((N-1)*D[i])) )

    # computation of the scores    
    Y = np.dot(Kc,An)
    
    return Y,An,D,var_explained
```

Independent Component Analysis (ICA) is a computational method for separating a multivariate signal into additive subcomponents that are maximally independent. It is commonly used in scenarios where the goal is to identify underlying factors or sources from mixed signals. Unlike PCA, which focuses on maximizing variance and finding orthogonal components, ICA aims to find components that are statistically independent from each other.

ICA can work better than PCA in the following scenarios:

1. **When the Data Contains Mixed Signals**: ICA excels in situations where the observed data are mixtures of unknown latent variables (or sources), and the goal is to separate these mixtures into their constituent sources.

2. **When Independence is More Important Than Variance**: In cases where signals or features are independent but not necessarily orthogonal or uncorrelated, ICA can separate these sources more effectively than PCA.

- **Scenario 1**: ICA appears to have separated the two classes along the first independent component reasonably well, suggesting that the underlying sources of the signals (features) are independent. If these sources correspond to our class separation, then ICA would be working effectively here.

- **Scenario 2**: the first independent component shows overlapping scores for Class 0 and Class 1, it suggests that this component has not captured features that are independent in a way that also corresponds to the class distinction. ICA would not be deemed successful in this scenario for the purpose of class separation.

- **Scenario 3**: The classes are distributed in a way that suggests there might be underlying independent processes or factors at play. If the ICA components correspond to these factors, ICA may work better than PCA by isolating these independent sources even if they don't correspond to maximum variance.

- **Scenario 4**: This scenario seems to depict a nonlinear relationship between classes, likely unsolvable by linear methods like PCA. If the circular components correspond to independent sources of variance in the data, ICA might be able to separate them effectively, which PCA could not.

In general, ICA will outperform PCA when the assumption of statistical independence is more appropriate for the data at hand than the assumption of uncorrelated components. It's important to note that ICA does not provide components in order of importance (like PCA does with variance), and all components are considered equally important. The success of ICA is also highly dependent on the assumption that the sources are statistically independent and that the mixing system is linear and instantaneous.

```python
def FastICALecture(X,N_Iter=3000,tol=1e-5,plot_evolution=0):
    '''
    Inputs: 
                   X: is a [d x N] matrix. Every column is an observation 
                   and every row is a feature.       
    
                   (Optional) N_Iter: maximum number of iterations
    
                   (Optional) delta: convergence criteria threshold
    
                   (Optional) plot_evolution: plot evolution of error
                   
     Outputs:      
                   S: [d x N] matrix.  Each column is an independent component 
                   of the centred and whitened input data X              
                   
                   W: [d x d] matrix. It is the demixing matrix. S = W*Xcw 
    
    '''
    
    # We choose G =-exp(-1/2t^2)
    # First derivative of G
    def g(t):
        res = t*np.exp(-(1/2)*(t**2))
        return res
    
    # Second derivative of G  
    def gp(t):
        res = (1 - t**2) * np.exp(-(1/2) * (t**2))
        return res
    
    # Size of X
    d,N = X.shape 
        
    # Compute sample mean
    mu = np.mean(X,axis=1)
    
    # Center data
    Xc = X - mu[:, np.newaxis]

    # Compute covariance matrix
    C = np.cov(Xc, rowvar=True)
    
    # Whiten data
    Xcw = np.dot(np.linalg.inv(np.linalg.cholesky(C)).T, Xc)
    
    # check if are whitened
    if np.sum(np.eye(d) - np.abs(np.cov(Xcw)))>1e-10:
        raise NameError('Your whitening transformation does not work...')
    
    # Initialize W
    W = ortho_group.rvs(d) # random orthogonal matrix 
    
    # delta evolution
    k = 0
    delta = np.inf
    evolutionDelta=[]
    
    while delta > tol and k < N_Iter:
    
        k = k + 1
        W_old = np.copy(W)
        
        # Update rule for W
        Wp = np.dot(g(np.dot(W, Xcw)), Xcw.T) - np.diag(np.dot(gp(np.dot(W, Xcw)), np.ones(N))) @ W
        W = np.linalg.qr(Wp)[0]  # Enforce orthogonality using QR decomposition
        # W = LA.inv( np.sqrt( Wp @ Wp.T ) ) @ Wp # Enforce orthogonality using normalization

        if np.sum(np.eye(d) - np.abs(np.dot(W,W.T)))>1e-10:
            raise NameError('W should be an orthogonal matrix. Check the computations')
                 
        delta = 1 - np.min( np.abs( np.diag( np.dot(W.T,W_old) ) ) )
        evolutionDelta.append(delta)
        
        if k == 1 or k % 100 == 0:
            print('Iteration ICA number ', k, ' out of ', N_Iter , ', delta = ', delta)
     
        
    if k == N_Iter:
        print('Maximum number of iterations reached ! delta = ', delta)
    else:
        print('Convergence achieved ( delta = ', delta, ') in ', k, ' iterations')


    # Independent components
    S = np.dot(W, Xcw)
            
    if plot_evolution == 1:
        plt.figure(figsize=(6, 6))
        plt.plot(range(k),evolutionDelta,'bx--', linewidth=4, markersize=12)  
        plt.title('Evolution of error - ICA')
        plt.show()
       
    return S, W, Xcw
```

# II - 3) ICA (optional part)

In the next section you will evaluate ICA. Every image $x_i$ can be seen as a linear combination of basis images. ICA can be used in two different ways for face recognition. We can look for a set of statistically independent basis images $s_j$ (first architecture) or for a set of statistically independent coefficients $a_{j}$ (second architecture).

In the first architecture, we compute $X'=A'S'$, where every row of $X'$ is an image and the columns are pixels. Images are considered as random variables and we look for a set of statistically independent basis images contained in the rows of $S'$.

In the second architecture, we transpose the previous setting computing $X''=A''S''$, where every column of $X''$ is an image and rows are pixels. In this case, we consider the pixels as random variables and we look for a set of statistically independent coefficients contained in the rows of $S$ and a set of basis images in the columns of $A$.

Instead than using the original training data $X$ as input matrix, we are going to use the eigenvectors (first architecture) or the scores (second architecture) computed with PCA, namely $Y=XL$ (same notation as in the slides of the lecture). In this way, we reduce the computational time since the number of eigenvectors that account for 99\% of the variance of the training images (columns of $L$) is definitely lower than the number of pixels (columns of $X$). If you want, you can of course use the original data but it will take much more time to converge.

For the first architecture we will use $L^T$ as input matrix. In fact, we can notice that the PCA approximation of the matrix $X_{train}$, containing an image in every row, can be written as $\tilde{X} = YL^T$. If we use $L^T$ as input in the ICA algorithm we obtain $L^T=AS$, thus it follows that $\tilde{X}=YW^TS$ (since $A=W^{-1}=W^T$). The basis images are contained in the rows of $S$ and the coefficients used for evaluating the performance are instead contained in the rows of $Y_{train}W^T$ for the training set and in $Y_{test}W^T$ for the test set.

For the second architecture, we will instead use $Y^T$ as input matrix thus obtaining $Y^T=AS$. Remember that in the second architecture we want to apply the ICA algorithm to the transpose of $X_{train}$. We can notice that, given the PCA transformation $Y=XL$, one can write $X \approx YL^T$ which entails $X^T \approx LY^T=LAS=LW^TS$. The columns of $LW^T$ contain the basis images whereas the columns of $S$ contain the statistically independent coefficients used to test the performance of the algorithm. The coefficients for the test set are in the columns of $S_{test}=W_{train}Y_{test}^T$.
